```python
import pymysql
import pandas as pd

host='127.0.0.1'
port="3306"
user='root'
password=''
database='classicmodels'

conn=pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')
def run(sql):
    df=pd.read_sql_query(sql,conn)
    return df
```

### One to many relationship¶
1. Report the account representative for each customer.

```python
run("""
select
    customername
from 
    customers c
inner join
    employees e on c.salesrepemployeenumber=e.employeenumber""")
```

2. Report total payments for Atelier graphique.

```python
run("""
select
    customernumber,
    sum(amount) as total_pembayaran
from 
    customers
inner join
    payments using (customernumber)
where customername='Atelier graphique'""")
```

3. Report the total payments by date

```python
run("""
select
    paymentdate,
    sum(amount)
from
    payments
group by paymentdate""")
```

4. Report the products that have not been sold

```python
run("""
select p.productcode
from 
    products p
left join 
    orderdetails o
on p.productcode=o.productcode
where 
    o.productcode is null""")
```

5. List the amount paid by each customer

```python
run("""
select
    c.customernumber,
    customername,
    sum(amount) as total
from
    customers c
inner join
    payments p
on c.customernumber=p.customernumber
group by customernumber""")
```

6. How many orders have been placed by Herkku Gifts?

```python
run("""
select 
    sum(quantityordered) as total_pesanan
from 
    orderdetails o1
        inner join
orders o2
    on o1.ordernumber=o2.ordernumber
where customernumber='167'
""")
```

7. Who are the employees in Boston?

```python
run("""
select
    concat(e.firstname,' ',e.lastname) as pegawai_di_boston
from
    employees e
inner join
    customers c
on e.employeenumber=c.salesrepemployeenumber
where city='boston'""")
```

8. Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.

```python
run("""
select
    customername,
    sum(amount) as total
from 
    customers c
inner join
    payments p on c.customernumber=p.customernumber
group by p.customernumber
having total>100000
order by total desc
    """)
```

9. List the value of 'On Hold' orders.

```python
run("""
select ordernumber
from orders
where 
status like '%on hold' or comments like '%on hold%'""")
```

10. Report the number of orders 'On Hold' for each customer

```python
run("""
select c.customername,
    o.ordernumber,
    status,
    comments
from 
    customers c
inner join 
    orders o using (customernumber)
where status like '%on hold' or comments like '%on hold%'
    """)
```

### Many to many relationship

1. List products sold by order date. 

```python
run("""
select 
    orderdate,
    productcode
from 
    orders
inner join
    orderdetails using (ordernumber)
""")
```

2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

```python
run("""
select orderdate
from orderdetails
orderdetails
inner join 
orders
using (ordernumber)
where productcode='S18_1097'
order by orderdate desc""")
```

3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?

```python
run("""
select 
    customername, 
    sum(quantityordered*priceeach) total
from 
    orderdetails c
left join
    orders o
using (ordernumber)
inner join
    customers
    using (customernumber)
group by ordernumber
having total>25000
""")
```

4. Are there any products that appear on all orders?

```python
run("""
select productcode
from 
(select count(ordernumber) as jumlah_order1 from orders) t1
inner join
(select productcode, count(ordernumber) as jumlah_order2 from orderdetails
group by productcode) t2
on t1.jumlah_order1=t2.jumlah_order2
""")
      
```

5. List the names of products sold at less than 80% of the MSRP.

```python
run("""
select productname, msrp, priceeach
from products 
inner join orderdetails using(productcode)
where priceeach<0.8*msrp""")
```

6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)

```python
run("""
select productname, buyprice, msrp, priceeach
from 
products 
inner join
orderdetails using (productcode)
where 
((priceeach-buyprice)*100/buyprice)>=100""")
```

7. List the products ordered on a Monday.

```python
run("""
select productname, dayname(orderdate) as day,
orderdate
from
products
inner join orderdetails using (productcode)
inner join
orders using (ordernumber)
where dayname(orderdate)='monday'""")
```

8. What is the quantity on hand for products listed on 'On Hold' orders?

```python
run("""
select 
productname,
count(productname) as total_produk
from
products inner join orderdetails using (productcode)
inner join orders using (ordernumber)
where status like '%on hold' or comments like '%on hold%'""")
```
